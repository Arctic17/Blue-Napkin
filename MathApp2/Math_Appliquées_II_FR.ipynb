# Packages

Quatre packages importants vont nous être utile

1.   **numpy** Pour effectuer du calcul numérique.
2.   **pandas** Pour créer et gérer des dataframes.
3.   **matplotlib.pyplot** Pour créer des figures.
4.    **seaborn** Pour améliorer le rendu des figures et importer des jeux de données.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Data frames

Le package seaborn permet d'importer une grande quantités de jeux de données différents. Nous allons importe le jeu de données **planets**.

```python
# Load Planets dataset
planets = sns.load_dataset("planets")
```

Pour avoir un aperçu d'un dataframe, on peut utiliser la commande **head()**.

```python
planets.head()
```

Le nombre de ligne et de colonne d'un dataframe peut être obtenu à l'aide de la commande **shape**.

```python
planets.shape
```

Pour accéder à **une colonne** en particulier, on écrit le nom de la colonne **entre crochets [ ]**.

```python
planets["mass"]
```

De manière similaire, on peut utiliser **.nom_de_colonne**.

```python
planets.year
```

Pour accéder à **une ligne** en particulier, on utilise la commande **loc[ ]**.

```python
planets.loc[18]
```

La commande **loc[ ]** permet aussi d'avoir certaines ligne d'une certaine colonne.

```python
planets.loc[10:15, "number"]
```

La commande **iloc[ ]** permet d'obtenir certaines **lignes** et certaines **colonnes** d'un dataframe.

```python
planets.iloc[10:15, 0:3]
```

Pour trier un dataframe selon une certaine colonne, on utilise la commande **sort_values**.

```python
planets.sort_values(by=['year'], ascending=True)
```

On peut aussi trier un dataframe à l'aide d'opérateurs logiques.

```python
planets[planets["year"] > 21]
```

```python
planets[planets["year"] == 2000]
```

# Barplot

```python
# Import the Planets dataset from seaborn's built-in datasets
planets = sns.load_dataset("planets")

# Create a figure and a single subplot with a specific size
ax = plt.subplots(1, 1, figsize=(10, 5))

# Set visual style parameters
sns.set_palette("colorblind")       # Use a colorblind-friendly palette
sns.set_context("talk")             # Increase font sizes for presentations
sns.set_style("white")              # Use a clean white background style

# Create a count plot for exoplanets discovered after 2004
# 'x' corresponds to the discovery year, 'hue' adds color distinction by year
count = sns.countplot(
    x='year',
    data=planets[planets["year"] > 2004],  # Filter to include only years after 2004
    hue="year",
    legend=False,                          # Remove the legend (it's redundant here)
    palette="dark:#0072B2"                        # Set bar color using a specific hex code
)

# Add a title and axis labels with specific font sizes
plt.title("Exoplanet discovery", fontsize=18)
count.set_xlabel("year", fontsize=16)
count.set_ylabel("number of exoplanets", fontsize=16)

# Set font size for y-axis tick labels
plt.yticks(fontsize=14)

# Display the plot
plt.show()
```

# Histogram

```python
# Import the Planets dataset from seaborn's built-in datasets
planets = sns.load_dataset("planets")

# Create a figure and a single subplot with a specific size
ax = plt.subplots(1, 1, figsize=(10, 5))

# Set visual style parameters
sns.set_palette("colorblind")       # Use a colorblind-friendly palette
sns.set_context("talk")             # Increase font sizes for presentations
sns.set_style("white")              # Use a clean white background style

# Create a histogram showing the distribution of exoplanet masses
# 'bins=25' sets the number of bins in the histogram
# The color is set using a specific hex code
count = sns.histplot(
    x='mass',
    data=planets,
    bins=25,
    color="#D55E00"
)

# Add a title and axis labels with specific font sizes
plt.title("Exoplanet discovery", fontsize=18)
count.set_xlabel("mass (in Jupiter masses)", fontsize=16)  # More explicit label
count.set_ylabel("number of exoplanets", fontsize=16)

# Display the plot
plt.show()
```

# Boxplot

```python
# Import the Planets dataset from seaborn's built-in datasets
planets = sns.load_dataset("planets")

# Create a figure and a single subplot with a specific size
ax = plt.subplots(1, 1, figsize=(10, 5))

# Set the title of the plot
plt.title("Exoplanet discovery", fontsize=18)

# Set visual style parameters
sns.set_palette("colorblind")       # Use a colorblind-friendly palette
sns.set_context("talk")             # Increase font sizes for presentations
sns.set_style("white")              # Use a clean white background style

# Create a boxplot to show the distribution of exoplanet attributes by year
# In this case, it will show the distribution of all numerical variables per year,
# which may include 'mass', 'orbital_period', etc., unless filtered beforehand.
sns.boxplot(
    data=planets,
    x="year",                       # Group data by discovery year
    color='#009E73',                # Set the box color using a hex code
    linewidth=3,                    # Thickness of boxplot lines
    width=0.3                       # Width of each box
)

# Display the plot
plt.show()
```

# Correlation

```python
# Import the Planets dataset from seaborn's built-in datasets
planets = sns.load_dataset("planets")

# Set the random seed for reproducibility
np.random.seed(42)

# Set visual style parameters
sns.set_palette("colorblind")       # Use a colorblind-friendly palette
sns.set_context("talk")             # Increase font sizes for presentations
sns.set_style("white")              # Use a clean white background style

# Number of data points to generate for each correlation
n = 100

# Generate synthetic data for different correlation scenarios

# Moderate negative correlation (approximately -0.65)
x1 = np.linspace(0, 10, n)
y1 = -0.028 * x1 + np.random.normal(0, 0.1, n)

# No correlation (approximately 0.0)
x2 = np.linspace(0, 10, n)
y2 = np.random.normal(0, 0.1, n)

# Strong positive correlation (approximately 0.95)
x3 = np.linspace(0, 10, n)
y3 = 0.1 * x3 + np.random.normal(0, 0.1, n)

# Create a figure with 3 subplots arranged in a single row
fig, axes = plt.subplots(1, 3, figsize=(14, 6))

# Plot the data for each correlation case using scatter plots
sns.scatterplot(x=x1, y=y1, ax=axes[0])
axes[0].set_title("Corr(x, y) ≈ -0.65")

sns.scatterplot(x=x2, y=y2, ax=axes[1])
axes[1].set_title("Corr(x, y) ≈ 0.0")

sns.scatterplot(x=x3, y=y3, ax=axes[2])
axes[2].set_title("Corr(x, y) ≈ 0.95")

# Improve layout and spacing
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Display the plots
plt.show()
```

# Linear regression

```python
# Import the Planets dataset from seaborn's built-in datasets
planets = sns.load_dataset("planets")

# Create a figure and a single subplot with a specific size
ax = plt.subplots(1, 1, figsize=(10, 5))

# Set the title of the plot
plt.title("Linear regression", fontsize=18)

# Set visual style parameters
sns.set_palette("colorblind")       # Use a colorblind-friendly palette
sns.set_context("talk")             # Increase font sizes for presentations
sns.set_style("white")              # Use a clean white background style

# Define the data
x = [1, 2, 3, 4]  # Explanatory variable
y = [2, 3, 5, 4]  # Target variable

# Plot the scatter points and the regression line
sns.regplot(
    x=x,
    y=y,
    ci=None,                            # No confidence interval
    scatter_kws={'color': "#D55E00", 's': 100},  # Customize point color and size
    line_kws={'color': '#009E73', 'lw': 3}       # Customize regression line color and width
)

# Customize axis labels and limits
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$y$", fontsize=16)
plt.xlim(0.5, 4.5)
plt.ylim(1, 5.5)

# Display the plot
plt.show()
```
